1. Tentukan Library

```python
import numpy as np
import pandas as pd 
from sklearn.preprocessing import StandardScaler 
from sklearn.model_selection import train_test_split 
from sklearn import svm 
from sklearn.metrics import accuracy_score
```

2. Load Dataset

```python
diabetes_dataset = pd.read_csv('diabetes.csv')
```

```python
diabetes_dataset.head(10)
```

```python
diabetes_dataset.shape
```

```python
diabetes_dataset['Outcome'].value_counts()
```

```python
# memisahkan data dan label 
X = diabetes_dataset.drop(columns='Outcome',axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
Y.head(10)
```

3. Standarisasi Data (Transformasi)

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standrize_data = scaler.transform(X)
```

```python
print(standrize_data)
```

```python
X = standrize_data 
Y = diabetes_dataset['Outcome']
```

```python
print(Y)
```

4. Memisahkan Data Training dan Data Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

5. Melatih data menggunakan algoritma SVM

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train, Y_train)
```

6. Membuat model evaluasi untuk mengukut tingkat akurasi

```python
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)
```

7. Membuat model prediksi


```python
input_data = (6, 148, 72, 35, 0, 33.6, 0.627, 50)
input_data_as_numpy_array = np.array(input_data)
input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshape)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('Tidak terkena diabetes')
else :
  print('Pasien Terkena diabetes')
```
